```python
# get sapecific

import requests
from bs4 import BeautifulSoup

data= requests.get('https://quotes.toscrape.com/')
soup= BeautifulSoup(data.text,'html.parser')

quotes=[]
for quote in soup.find_all('span',class_='text'):
  quotes.append(quote.text[1:-1])

print(quotes)
```

```python
#scrapping Scope in details

import requests
import pandas as pd
from bs4 import BeautifulSoup

source = requests.get('https://quotes.toscrape.com/')
res= BeautifulSoup(source.text,'html.parser')

data=[]
for sp in res.find_all('div',class_='quote' ):
  quotes= sp.find('span', class_='text')
  author= sp.find('small',class_='author')
  author_id= sp.find('a').get('href')

  tags=[]
  for tag in sp.find_all('a', class_='tag'):
   tags.append(tag.text)

  ','.join(tags)

  data.append([quotes.text[1:-1],author.text,author_id,tags])

print(data)

df=pd.DataFrame(data,columns=['quotes','author','author_id','tags'])
df.to_csv('quotes.csv')




```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm   #tqdm is used to show a progress bar for long-running loops.


dataset=[]
for i in tqdm(range(1,11)):
  link ='https://quotes.toscrape.com/page/'+ str(i)
  res  = requests.get(link)
  data = BeautifulSoup(res.text,'html.parser')


  for soup in data.find_all('div',class_='quote'):
    quote=soup.find('span',class_='text').text[1:-1]
    author=soup.find('small',class_='author').text
    author_id=soup.find('a').get('href')
    tags=[]
    for tag in soup.find_all('a',class_='tag'):
        tags.append(tag.text)

    dataset.append([quote,author,author_id,tags])

df=pd.DataFrame(dataset,columns=['Quotes','Author','Author_id','Tags'])

print(df)
```
